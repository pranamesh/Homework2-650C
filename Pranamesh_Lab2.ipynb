```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(2,4,0.1)
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.tan(x)
f, axarr = plt.subplots(2, 2, sharex=True,sharey=True)
axarr[0, 0].plot(x, y1)
axarr[0, 0].set_title('y = cos(x)')
axarr[0, 1].plot(x, y2)
axarr[0, 1].set_title('y = sin(x)')
axarr[1, 0].plot(x, y3)
axarr[1, 0].set_title('y = tan(x)')
axarr[1, 1].plot(x, x)
axarr[1, 1].set_title('y = x')
```

```python
# 2b
plt.hold(True)
plt.plot(x, y1, 'k', linewidth = 4, label="cos(x)")
plt.plot(x, y2, 'g', linewidth = 4, label="sin(x)")
plt.plot(x, y3, 'r', linewidth = 2, label="tan(x)")
plt.plot(x, x, 'b', linewidth = 2, label="x")
legend = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# 2c
import csv
with open('Temperature.csv', 'r') as f:
    reader = csv.reader(f)
    temp_data = list(reader)
```

```python
temp = []
date = []
for i in range(1,len(temp_data)):
    temp.append(int(temp_data[i][1]))
    date.append(int(temp_data[i][0]))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(date,temp)
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Day vs Temperature')
```
